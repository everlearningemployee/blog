```python
#hide
import matplotlib.pyplot as plt
from matplotlib import font_manager
plt.rcParams["font.family"] = 'gulim'
plt.rcParams["figure.figsize"] = (15,5)

import pandas as pd
```

```python
csv_url = 'https://covid19.who.int/WHO-COVID-19-global-data.csv'
df = pd.read_csv(csv_url).rename(columns=lambda x: x.strip())
df.Date_reported = pd.to_datetime(df.Date_reported)

df_country = df[(df.Date_reported > '2020') & (df.Country_code == 'US')]
df_country = df_country.set_index('Date_reported')
df_country.New_cases.plot()
```
